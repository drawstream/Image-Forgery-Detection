```python
import numpy as np # linear algebra
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import math
import tensorflow as tf
import time
import os
import cv2
from sklearn.metrics import accuracy_score
import time
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
model = ResNet50(weights='imagenet', include_top=False)
model.summary()
```

```python
def feature_extraction(path,model=model):
# Reference from https://franky07724-57962.medium.com/using-keras-pre-trained-models-for-feature-extraction-in-image-clustering-a142c6cdf5b1    
    img = image.load_img(path, target_size=(224, 224)) # reshapes given image into 224x224 using bilinear interpolation
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    feature = model.predict(img_data, verbose=1)
    return feature
```

```python
PATH = r"C:\Users\vish\COFOMOD_v2\\"
file_formats = [".png", ".jpg", ".jpeg"]  # image file formats
```

```python
paths = []
images = []

for x in os.listdir(PATH):
    istrue = np.array([x.endswith(file_format)
                       for file_format in file_formats])
    if np.any(istrue):
        # detecting png,jpg and jpeg files which are most common image files (in dataset it has only these filetypes)
        paths.append(PATH+x)
        images.append(x)
```

```python
x = []
for index,path in enumerate(paths):
    stre = images[index].split("_")
    if (len(stre[1]) != 1 and not images[index].endswith('_O.png') and not images[index].endswith('_F.png')):
        # we are skipping bit mask images as they are not images that we have to consider
        pass
    else:
        resnet_feature = feature_extraction(path)
        x.append(resnet_feature) # feature matrix
```

```python
X = np.array(x)
y = []

for index,path in enumerate(paths):
    # detection using pre computed Gabor filters
    stre = images[index].split("_")
    if (len(stre[1]) != 1 and not images[index].endswith('_O.png') and not images[index].endswith('_F.png')):
        # we are skipping bit mask images as they are not images that we have to consider
        pass
    else:
        real_detected = 0 
        if(stre[1] == "F" or stre[1] == "F.png" or stre[1] == "F.jpg" or stre[1] == "F.jpeg" ):
            real_detected = 1
        y.append(real_detected) # either one or 0 classifier
y = np.array(y)
```

```python
p, q, r, s, t = X.shape 
print(X.shape)
```

```python
X_new = np.zeros((p, q*r*s*t))
for i in range(X.shape[0]):
    X_new[i] = X[i].flatten()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2)
clas = SVC(kernel= 'linear', gamma='auto')
clas.fit(X_train, y_train)
```

```python
CMFD = clas.predict(X_test)
print("Accuracy is:-",accuracy_score(y_test,CMFD))
```

```python
random_forest = RandomForestClassifier().fit(X_train, y_train)
```

```python
random_forest_output = random_forest.predict(X_test)
print("Accuracy is:-",accuracy_score(y_test,random_forest_output))
```
