```python
import numpy as np
import os
import PIL
from sklearn.svm import SVC
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
tf.config.list_physical_devices('GPU')
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

Taking the VGG16 Implementation that was trained on Imagenet Database from Tensorflow.

```python
model = VGG16(weights='imagenet', include_top=False)
model.summary()

PATH = r"C:\Users\vish\COFOMOD_v2\\"
file_formats = [".png", ".jpg", ".jpeg"]
```

```python
paths = []
images = []

for x in os.listdir(PATH):
    istrue = np.array([x.endswith(file_format)
                       for file_format in file_formats])
    if np.any(istrue):
        paths.append(PATH+x)
        images.append(x)
```

```python
x = []
y = []

for index,path in enumerate(paths):
    # detection using pre computed Gabor filters
    stre = images[index].split("_")
    if (len(stre[1]) != 1 and not images[index].endswith('_O.png') and not images[index].endswith('_F.png')):
        # we are skipping bit mask images as they are not images that we have to consider
        pass
    else:
        img = image.load_img(path, target_size=(224, 224))
        img_data = image.img_to_array(img)
        img_data = np.expand_dims(img_data, axis=0)
        img_data = preprocess_input(img_data)
        vgg16_feature = model.predict(img_data, verbose=1)
        x.append(vgg16_feature)
        real_detected = 0
        if(stre[1]=="F" or stre[1]=="F.png" or stre[1]=="F.jpg" or stre[1]=="F.jpeg"):
            real_detected=1
        y.append(real_detected)
```

```python
X = np.array(x)
y = np.array(y)

print(X.shape)
print(y.shape)
```

```python
X_new = np.zeros((10000, 49*512))
for i in range(X.shape[0]):
    X_new[i] = X[i].flatten()

X_new.shape
```

```python
print(np.unique(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3)
clas = SVC(kernel= 'linear', gamma='auto')
clas.fit(X_train, y_train)
```

```python
CMFD = clas.predict(X_test)
print("Accuracy is:-",accuracy_score(y_test,CMFD))
```

```python
clf = RandomForestClassifier()  # Not using hyperparameters as CV would take up a lot of time (n_estimators)
clf.fit(X_train, y_train)
```

```python
CMFD3 = clf.predict(X_test)
print("Accuracy3 is :- ", accuracy_score(y_test,CMFD3))
```
