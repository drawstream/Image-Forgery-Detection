```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
import tensorflow as tf
import time
import sys
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
model = InceptionV3(weights='imagenet')
model.summary()

PATH = r"C:\Users\vish\COFOMOD_v2\\"
file_formats = [".png", ".jpg", ".jpeg"]  # image file formats
```

```python
paths = []
images = []

for x in os.listdir(PATH):
    istrue = np.array([x.endswith(file_format)
                       for file_format in file_formats])
    # print(istrue)
    if np.any(istrue):
        # detecting png,jpg and jpeg files which are most common image files (in dataset it has only these filetypes)
        paths.append(PATH+x)
        images.append(x)
```

```python
x = []
y = []
i=0
for index,path in enumerate(paths):
    # detection using pre computed Gabor filters
    stre = images[index].split("_")
    if (len(stre[1]) != 1 and not images[index].endswith('_O.png') and not images[index].endswith('_F.png')):
        # we are skipping bit mask images as they are not images that we have to consider
        pass
    else:
        img = image.load_img(path, target_size=(299, 299))
        img_data = image.img_to_array(img)
        img_data = np.expand_dims(img_data, axis=0)
        img_data = preprocess_input(img_data)
        googlenet_feature = model.predict(img_data, verbose=1)
        x.append(googlenet_feature)
        if stre[1] == "F" or stre[1] == "F.png" or stre[1] == "F.jpg" or stre[1] == "F.jpeg":
          real_detected = 1
        else:
          real_detected = 0
        # real_detected = 1 if stre[1] == "F" or "F.png" or "F.jpg" or "F.jpeg" else 0
        y.append(real_detected) # either one or 0 classifier
        if i % 100 == 0:
          print(i)
        i+=1

X = np.array(x)
y = np.array(y)
```

```python
print(X.shape)
print(y.shape)
print(y)
```

```python
p, q, r = X.shape 
```

```python
X_new = np.zeros((p, q*r))
for i in range(X.shape[0]):
  X_new[i] = X[i].flatten()
```

```python
X_new.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2)
```

```python
clas = SVC(kernel= 'linear', gamma='auto')
clas.fit(X_train, y_train)
```

```python
CMFD = clas.predict(X_test)
print("Accuracy is:-",accuracy_score(y_test,CMFD))
```

```python
random_forest = RandomForestClassifier().fit(X_train, y_train)
```

```python
random_forest_output = random_forest.predict(X_test)
print("Accuracy is:-",accuracy_score(y_test,random_forest_output))
```
